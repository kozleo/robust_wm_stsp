```python
import os

import matplotlib.pyplot as plt
import numpy as np
import xarray as xr

from stabnhp.utils import xr_load, get_results_path, get_delay_times, get_session_ids
```

```python
results_path = get_results_path()
```

```python
def create_summary_storage():
    sessions = get_session_ids()
    n_sessions = sessions.size
    
    delays = get_delay_times()
    n_delays = delays.size
    
    kernel_widths = np.array([5, 10, 15])
    n_kernel_widths = kernel_widths.size
    
    bin_sizes = np.array([20, 50, 100])
    n_bin_sizes = bin_sizes.size
    
    epochs = np.array(['sample', 'pre_test'])
    n_epochs = epochs.size
    

    res = xr.DataArray(
        np.zeros((n_sessions, n_delays, n_kernel_widths, n_bin_sizes, n_epochs)),
        dims=["session", "delay", "kernel_width", "bin_size", "epoch"],
        coords={"session": sessions, "delay": delays, "kernel_width": kernel_widths, 
                "bin_size": bin_sizes, "epoch": epochs},
    )

    return res
```

```python
summary_res = create_summary_storage()
```

```python
kws = summary_res.coords['kernel_width'].data
bss = summary_res.coords['bin_size'].data
```

```python
for kw in kws:
    for bs in bss:
        filename = f'decode_sample_{kw}_{bs}.nc'
        filepath = os.path.join(results_path, filename)
        results = xr_load(filepath)

        sessions = results.session.data
        delays = summary_res.delay.data
        time = results.timepts

        n_sessions = sessions.size
        n_delays = delays.size

        samp_scores = np.zeros((sessions.size, delays.size))
        pret_scores = np.zeros((sessions.size, delays.size))

        for s, session in enumerate(sessions):
            for d, delay in enumerate(delays):
                test_start = ((delay*1000)+500) * 1e-3

                sample_filt = ((time > 0) & (time < 0.5))
                pre_test_filt = ((time > test_start - 0.1) & (time < test_start - 0.0))

                # use no-distractor trials
                data = results.loc[session,delay,0,:]

                samp_scores[s,d] = data[sample_filt].mean()

                pret_scores[s,d] = data[pre_test_filt].mean()

        summary_res.loc[:,:,kw,bs,'sample'] = samp_scores
        summary_res.loc[:,:,kw,bs,'pre_test'] = pret_scores
```

```python
from scipy.stats import wilcoxon
```

```python
plt.style.use('../sci_style.py')
fig, ax = plt.subplots(3,3, figsize=(10,10), sharex=True, sharey=True)
pvals = []
for i, bs in enumerate(bss):
    for j, kw in enumerate(kws):
        
        s_data = summary_res.loc[:,:,kw,bs,'sample'].data
        s_data = s_data.flatten()
        
        t_data = summary_res.loc[:,:,kw,bs,'pre_test'].data
        t_data = t_data.flatten()
        
        test = wilcoxon(s_data, t_data)
        pval = test.pvalue
        pvals.append(pval)
        
        ax[i,j].scatter(np.repeat(0,s_data.size),s_data, color='b')
        ax[i,j].scatter(np.repeat(1,t_data.size),t_data, color='r')
        for q in range(s_data.size):
            ax[i,j].plot([s_data[q], t_data[q]], color='k', alpha=0.2)
        ax[i,j].axhline(0.133, linestyle='--', color='cyan')
    
        if i == 0:
            ax[i,j].set_title(f'kernel width = {kw}', fontsize=16)
        if j == 0:
            ax[i,j].set_ylabel(f'bin size = {bs}\n Accuracy', fontsize=18)
        ax[i,j].set_xlim([-0.5, 1.5])
        ax[i,j].set_xticks([0,1])
        ax[i,j].set_xticklabels(['Sample', 'Pre-Test'], fontsize=16)
plt.tight_layout()
plt.savefig('figures/decode_sample_hyper.png', dpi=300)
```

```python
from statsmodels.stats.multitest import fdrcorrection
pvals_c = fdrcorrection(pvals, alpha=0.05)
```
